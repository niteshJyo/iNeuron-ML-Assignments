### In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
```

```python
data = pd.read_csv("iris.csv")
data = data.drop(["Id"], axis = 1)
```

```python
data
```

```python
# fig 01

fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111,projection='3d')
ax.set_xlabel('SepalLength', fontsize = 15)
ax.set_ylabel('SepalWidth', fontsize = 15)
ax.set_zlabel('PetalLength', fontsize = 15)
ax.set_title('IRIS', fontsize = 20)

Species = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for species, color in zip(Species,colors):
    indicesToKeep = data['Species'] == species
    ax.scatter(data.loc[indicesToKeep, 'SepalLengthCm']
               , data.loc[indicesToKeep, 'SepalWidthCm']
               , data.loc[indicesToKeep, 'PetalLengthCm']
               , c = color
               , s = 50)
    
ax.legend(Species)
ax.grid()
plt.show()
```
